## Scale down the numerical data for readability

```python
# convert the population to millions for better understanding
world_bank_population_data_v3 = world_bank_population_data_v2.copy()

world_bank_population_data_v3[world_bank_population_data_v3.columns[1:]] = (
    world_bank_population_data_v3[world_bank_population_data_v3.columns[1:]] / 1_000_000
)
```

```python
# sample data
world_bank_population_data_v3.head()
```

- Some entries in the dataset represent **groups of countries**, which can lead to duplication since the population of individual countries is also included separately.  
  - For instance, `Sub-Saharan Africa (IDA & IBRD countries)` aggregates multiple European nations, such as Poland, Czech Republic (Czechia), and Slovakia (Slovak Republic).  

- The dataset also includes **country groups with ambiguous names**, such as **Small States** (countries with a population of 1.5 million or fewer) and **Upper Middle Income** (countries classified by a Gross National Income (GNI) per capita between $4,516 and $14,005).  
  - Since individual countries from these groups are already present in the dataset, keeping these entries may introduce redundancy and impact efficiency.  

- To address this, we will create **two separate dataframes**one including these country groups and one excluding themsince outright deletion of data is not an ideal approach.

### Filter Individual and Regional & Group-Based Countries

```python
# List of regional & group-based countries to be excluded
extra_countries = ['Africa Eastern and Southern','North America','Euro area','High income','South Asia','World','Arab World','Africa Western and Central','Latin America & Caribbean','Central Europe and the Baltics', 'Caribbean small states', 'East Asia & Pacific (excluding high income)', 'Early-demographic dividend', 'East Asia & Pacific','Europe & Central Asia (excluding high income)','Europe & Central Asia','Fragile and conflict affected situations','European Union','Heavily indebted poor countries (HIPC)','IBRD only','IDA & IBRD total','IDA total','IDA blend','IDA only','Latin America & Caribbean (excluding high income)','Least developed countries: UN classification','Low income','Lower middle income','Low & middle income','Late-demographic dividend','Middle East & North Africa','Middle income','Middle East & North Africa (excluding high income)','OECD members','Other small states','Pre-demographic dividend','Pacific island small states','Post-demographic dividend','Sub-Saharan Africa (excluding high income)','Sub-Saharan Africa','Small states','East Asia & Pacific (IDA & IBRD countries)','Europe & Central Asia (IDA & IBRD countries)','Latin America & the Caribbean (IDA & IBRD countries)','Middle East & North Africa (IDA & IBRD countries)','South Asia (IDA & IBRD)','Sub-Saharan Africa (IDA & IBRD countries)','Upper middle income']

# create a new dataframe excluding the selected countires
world_bank_population_data_v4 = world_bank_population_data_v3[~world_bank_population_data_v3['Country Name'].isin(extra_countries)]

# reset the index
world_bank_population_data_v4 = world_bank_population_data_v4.reset_index(drop=True)

# display sample data with excluded countries
world_bank_population_data_v4.head()
```

```python
# create a new dataframe only with the regional & group-based countries
world_bank_population_data_v5 = world_bank_population_data_v3[world_bank_population_data_v3['Country Name'].isin(extra_countries)]

# reset the index
world_bank_population_data_v5 = world_bank_population_data_v5.reset_index(drop=True)

# display sample data with only included countires
world_bank_population_data_v5.head()
```

# Visualization

## Individual Countries

### Global Population Growth Over Years (1960 - 2023)

```python
# Convert year columns to a single 'Year' column with corresponding values
df_melted_v1 = world_bank_population_data_v4.melt(
    id_vars=[world_bank_population_data_v4.columns[0]],
    var_name="Year",    # New column name for years
    value_name="Population in millions"  # New column name for the population values
)

# Compute total population for each year
df_total_population = df_melted_v1.groupby('Year', as_index=False)['Population in millions'].sum().round(2)

# Print the new DataFrame
df_total_population.head()
```

```python
# Set figure size
plt.figure(figsize=(15, 8))

# Create line plot
plt.plot(df_total_population['Year'], df_total_population['Population in millions'], marker='o', linestyle='-')

# Add title and labels
plt.title("Global Population Growth Over Years (1960-2023)", fontsize=14, loc='center')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Population (millions)", fontsize=12)
plt.xticks(rotation = 90)

# Show grid
plt.grid(True, linestyle='--', alpha=0.6)

# Show plot
plt.show()
```

### Global Population Growth Difference Over Time By Comparing Each Year (1960 - 2023)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert year columns to a single 'Year' column with corresponding values
df_melted_v2 = world_bank_population_data_v4.melt(
    id_vars=['Country Name'],
    var_name="Year",
    value_name="Population"
)

# Convert 'Year' to integer for proper plotting
df_melted_v2['Year'] = df_melted_v2['Year'].astype(int)

# Sort based on Country and Year
df_sorted = df_melted_v2.sort_values(by=['Country Name', 'Year'])

# Create a new column to store the population change across all countries
df_sorted['Population Change'] = df_sorted.groupby('Country Name')['Population'].diff().round(2)

# Sum population change across all countries for each year
global_population_change = df_sorted.groupby('Year')['Population Change'].sum().reset_index()

# Display first few rows
print(global_population_change.head())

# Matplotlib line plot for yearly population change
plt.figure(figsize=(10, 5))
plt.plot(global_population_change['Year'], global_population_change['Population Change'], marker='o', linestyle='-')

# Title and labels
plt.title("Global Population Growth Difference Over Time")
plt.xlabel("Year")
plt.ylabel("Yearly Population Growth (millions)")
plt.grid(True)

# Show the plot
plt.show()
```

### Top 5 Countries with Highest and Least Growth Difference Over Time By Comparing Each Year (1960 - 2023)

```python
# Convert Year into numeric format
df_sorted['Year'] = df_sorted['Year'].astype(int)

# Sum population change for each country by year
global_population_change = df_sorted.groupby(['Country Name', 'Year'])['Population Change'].sum().reset_index()

# Sort by Country and Year (ensuring top-to-bottom sorting for clarity)
df_countries = global_population_change.sort_values(by=['Population Change', 'Country Name', 'Year'], ascending=[False, True, True])

# List of top 5 countries with the highest population change difference
top_5_countries_growthDiff = df_countries.groupby('Country Name')['Population Change'].sum().nlargest(5).index.tolist()

# List of bottom 5 countries with the least population change difference
bottom_5_countries_growthDiff = df_countries.groupby('Country Name')['Population Change'].sum().nsmallest(5).index.tolist()

# Print top 5 and bottom 5 country names
print("Top 5 Countries with Highest Growth Difference:", top_5_countries_growthDiff)
print("Bottom 5 Countries with Least Growth Difference:", bottom_5_countries_growthDiff)

# Filter data for only these countries and retain Year
df_top5Diff = df_countries[df_countries['Country Name'].isin(top_5_countries_growthDiff)]
df_bottom5Diff = df_countries[df_countries['Country Name'].isin(bottom_5_countries_growthDiff)]

# Plot both top and bottom 5 countires
fig, axes = plt.subplots(2, 1, figsize=(15, 12), sharex=True)

# Plot Top 5 Countries
sns.lineplot(data=df_top5Diff, x="Year", y="Population Change", hue="Country Name", marker='o', ax=axes[0])
axes[0].set_title("Top 5 Countries - Yearly Population Growth", fontsize=14)
axes[0].set_ylabel("Population Change (millions)")
axes[0].legend(title="Country")

# Plot Bottom 5 Countries
sns.lineplot(data=df_bottom5Diff, x="Year", y="Population Change", hue="Country Name", marker='o', ax=axes[1])
axes[1].set_title("Bottom 5 Countries - Yearly Population Growth", fontsize=14)
axes[1].set_xlabel("Year")
axes[1].set_ylabel("Population Change (millions)")
axes[1].legend(title="Country")

plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.tight_layout()
plt.show()
```

## Population Growth in Different Countires Over Time (1960 - 2023)

```python
# Set Seaborn style for better aesthetics
sns.set_style("whitegrid")

# Compute the total population for each country and sort by highest to lowest
total_population = df_melted_v2.groupby("Country Name")["Population"].sum().sort_values(ascending=False)
sorted_countries = total_population.index.tolist()

# Create the plot
plt.figure(figsize=(12, 6))

# Loop through sorted countries and plot their data
for country in sorted_countries:
    country_data = df_melted_v2[df_melted_v2["Country Name"] == country]
    plt.plot(country_data["Year"], country_data["Population"], marker='o', label=country)

# Add title and labels
plt.title("Population Growth Over Time for Different Countries", fontsize=14, fontweight='bold')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Population (millions)", fontsize=12)
plt.xticks(rotation=45)  # Rotate x-axis labels for readability

# Add sorted legend
handles, labels = plt.gca().get_legend_handles_labels()
labels, handles = zip(*sorted(zip(labels, handles), key=lambda x: total_population[x[0]], reverse=True))
plt.legend(handles, labels, title="Country", loc='upper left', bbox_to_anchor=(1, 1))

# Show the plot
plt.show()
```

### Top 5 and Least 5 Populous Countries (1960 - 2023)

```python
# Find the top 5 and bottom 5 populated countries for the latest year (2023)
latest_year = 2023
df_latest = df_melted_v2[df_melted_v2['Year'] == latest_year]  # Filter for 2023

top_5_countries = df_latest.nlargest(5, 'Population')['Country Name'].tolist()
bottom_5_countries = df_latest.nsmallest(5, 'Population')['Country Name'].tolist()

# Print top 5 and bottom 5 country names
print("Top 5 Countries with Highest Population:", top_5_countries)
print("Bottom 5 Countries with Least Population:", bottom_5_countries)

# Filter data for only these countries
df_top5 = df_melted_v2[df_melted_v2['Country Name'].isin(top_5_countries)]
df_bottom5 = df_melted_v2[df_melted_v2['Country Name'].isin(bottom_5_countries)]

# Create subplots
fig, axes = plt.subplots(2, 1, figsize=(15, 12), sharex=True)

# Plot Top 5 Countries
sns.lineplot(data=df_top5, x="Year", y="Population", hue="Country Name", marker='o', ax=axes[0])
axes[0].set_title("Top 5 Highly Populated Countries Over Time", fontsize=14)
axes[0].set_ylabel("Population (millions)")
axes[0].legend(title="Country")

# Plot Bottom 5 Countries
sns.lineplot(data=df_bottom5, x="Year", y="Population", hue="Country Name", marker='o', ax=axes[1])
axes[1].set_title("Bottom 5 Least Populated Countries Over Time", fontsize=14)
axes[1].set_xlabel("Year")
axes[1].set_ylabel("Population (millions)")
axes[1].legend(title="Country")

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

# Grouped and Regional-Based Countries

### Global Population Growth for Regional-Based Countires Over Years (1960 - 2023)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
df_groupedCountries = world_bank_population_data_v5

# Convert year columns to a single 'Year' column with corresponding values
df_melted_v3 = df_groupedCountries.melt(
    id_vars=['Country Name'],      # Keep country names fixed
    var_name="Year",               # Convert column names (years) into a new column
    value_name="Population"        # Convert population values accordingly
)

# Convert 'Year' to integer for proper plotting
df_melted_v3['Year'] = df_melted_v3['Year'].astype(int)

# Set Seaborn style for better aesthetics
sns.set_style("whitegrid")

# Compute the total population for each country and sort by highest to lowest
total_population = df_melted_v3.groupby("Country Name")["Population"].sum().sort_values(ascending=False)
sorted_countries = total_population.index.tolist()

# Create the plot
plt.figure(figsize=(12, 6))

# Loop through sorted countries and plot their data
for country in sorted_countries:
    country_data = df_melted_v3[df_melted_v3["Country Name"] == country]
    plt.plot(country_data["Year"], country_data["Population"], marker='o', label=country)

# Add title and labels
plt.title("Population Growth Over Time for Different Countries", fontsize=14, fontweight='bold')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Population (millions)", fontsize=12)
plt.xticks(rotation=45)  # Rotate x-axis labels for readability

# Add sorted legend
handles, labels = plt.gca().get_legend_handles_labels()
labels, handles = zip(*sorted(zip(labels, handles), key=lambda x: total_population[x[0]], reverse=True))
plt.legend(handles, labels, title="Country", loc='upper left', bbox_to_anchor=(1, 1))

# Show the plot
plt.show()
```

### Top 5 and Least 5 Populous Reigonal-Based Countries (1960 - 2023)

```python
# Ensure Year column is integer for proper sorting
df_melted_v3['Year'] = df_melted_v3['Year'].astype(int)

# Select the latest year (2023) to determine top & bottom 5 countries
latest_year = 2023
df_latest = df_melted_v3[df_melted_v3['Year'] == latest_year]  # Filter data for 2023

# Get top 5 and bottom 5 populated countries based on 2023 data
top_5_countries = df_latest.nlargest(5, 'Population')['Country Name'].tolist()
bottom_5_countries = df_latest.nsmallest(5, 'Population')['Country Name'].tolist()

# Filter the dataset from 1960 to 2023 for these selected countries
df_top5 = df_melted_v3[df_melted_v3['Country Name'].isin(top_5_countries)]
df_bottom5 = df_melted_v3[df_melted_v3['Country Name'].isin(bottom_5_countries)]

# Create subplots
fig, axes = plt.subplots(2, 1, figsize=(15, 12), sharex=True)

# Plot Top 5 Countries over the period 1960-2023
sns.lineplot(data=df_top5, x="Year", y="Population", hue="Country Name", marker='o', ax=axes[0])
axes[0].set_title("Top 5 Highly Populated Countries (1960-2023)", fontsize=14)
axes[0].set_ylabel("Population (millions)")
axes[0].legend(title="Country")

# Plot Bottom 5 Countries over the period 1960-2023
sns.lineplot(data=df_bottom5, x="Year", y="Population", hue="Country Name", marker='o', ax=axes[1])
axes[1].set_title("Bottom 5 Least Populated Countries (1960-2023)", fontsize=14)
axes[1].set_xlabel("Year")
axes[1].set_ylabel("Population (millions)")
axes[1].legend(title="Country")

# Rotate x-axis labels for better readability
axes[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
